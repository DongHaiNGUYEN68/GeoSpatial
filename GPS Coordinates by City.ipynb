```python
def get_gps_coordinates(country_code, city):
    """
    Uses a geocoding API to retrieve GPS coordinates (latitude and longitude)
    for a given city and country code.  Handles rate limiting and User-Agent.

    Args:
        country_code (str): The two-letter ISO country code (e.g., "US", "JP", "FR").
        city (str): The name of the city (e.g., "Tokyo", "Paris", "London").

    Returns:
        tuple: A tuple containing the latitude and longitude as floats, or None
               if the API request fails or no results are found.  Returns (None, None)
               in case of error.
    """

    geocode_url = "https://nominatim.openstreetmap.org/search"

    params = {
        "country": country_code,
        "city": city,
        "format": "json",
        "limit": 1
    }

    headers = {
        "User-Agent": "Richemont Geocoder/1.0 (dong-hai.nguyen@richemont.com)"  # Replace with your email or app info
    }

    try:
        response = requests.get(geocode_url, params=params, headers=headers)
        response.raise_for_status()
        data = response.json()

        if data:
            latitude = float(data[0]["lat"])
            longitude = float(data[0]["lon"])
            return latitude, longitude
        else:
            print(f"No results found for {city}, {country_code}")
            return None, None

    except requests.exceptions.RequestException as e:
        print(f"API request failed: {e}")
        return None, None
    except (KeyError, ValueError) as e:
        print(f"Error parsing API response: {e}")
        return None, None
```

```python
# Example usage:
country_code = "JP"
city = "Tokyo"

latitude, longitude = get_gps_coordinates(country_code, city)

if latitude is not None and longitude is not None:
    print(f"{city},{city.upper()},{latitude},{longitude}")  # Output format as requested
else:
    print(f"Could not retrieve coordinates for {city}, {country_code}")
```
